2. H₀ (Null Hypothesis):

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

### Hypothetical Statement 1: Research Hypothesis

Null Hypothesis (H₀):
There is no significant difference between Tesla’s opening price and closing price.

Alternative Hypothesis (H₁):
There is a significant difference between Tesla’s opening price and closing price.


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_rel

# Perform Paired T-Test (since Open & Close are related)
t_stat, p_value = ttest_rel(df['Open'], df['Close'])

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The test performed is a Paired T-Test (also known as a dependent sample t-test).

##### Why did you choose the specific statistical test?

The Open and Close prices belong to the same trading day, so they are dependent (paired) observations.
A paired t-test is used to compare the mean difference between two related groups.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

### Hypothetical Statement 2: Research Hypothesis

Null Hypothesis (H₀):
Trading volume has no significant relationship with Tesla’s stock price changes.

Alternative Hypothesis (H₁):
Trading volume has a significant relationship with Tesla’s stock price changes.


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Create a new variable: Price Change
df['Price_Change'] = df['Close'] - df['Open']

# Perform Pearson Correlation Test
corr, p_value = pearsonr(df['Volume'], df['Price_Change'])

print("Correlation Coefficient:", corr)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value, a One-Sample T-Test was performed

##### Why did you choose the specific statistical test?

The One-Sample T-Test is used to determine whether the mean of a sample is significantly different from a known or hypothesized value. In this case, we tested whether the average daily return of Tesla stock is significantly different from 0.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

State Your Research Hypothesis

Null Hypothesis (H₀):
There is no significant relationship between Tesla's Closing Price and Trading Volume.

Alternative Hypothesis (H₁):
There is a significant relationship between Tesla's Closing Price and Trading Volume.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import pearsonr

# Perform Pearson Correlation Test
corr_coef, p_value = pearsonr(df['Close'], df['Volume'])

print("Correlation Coefficient:", corr_coef)
print("P-Value:", p_value)

# Conclusion
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis (H₀)")
    print("There is a significant relationship between Closing Price and Volume.")
else:
    print("Fail to Reject Null Hypothesis (H₀)")
    print("There is no significant relationship between Closing Price and Volume.")
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test was performed to obtain the p-value. This test measures the strength and direction of the linear relationship between Tesla's Closing Price and Trading Volume.

##### Why did you choose the specific statistical test?

The Pearson Correlation Test was selected because it is suitable for measuring the relationship between two continuous numerical variables. Since the hypothesis aimed to determine whether Tesla's Closing Price and Trading Volume are significantly related, this test was used to calculate both the correlation coefficient and the p-value. This helps identify the strength and statistical significance of the relationship.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# ============================================================
# Detecting Outliers Using IQR Method
# ============================================================

# Select numerical columns
numerical_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for col in numerical_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} outliers")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were identified using the IQR method and treated using capping (Winsorization). This technique was chosen because it minimizes the influence of extreme values while retaining all observations, which is important in stock market data where large price movements may represent real market behavior.

### 4. Textual Data Preprocessing

#### 1. Expand Contraction

````python
# Expand Contraction
# ============================================================
contractions = {
    "can't": "cannot",
    "won't": "will not",
    "don't": "do not",
    "didn't": "did not",
    "isn't": "is not",
    "aren't": "are not",
    "I'm": "I am",
    "it's": "it is",
    "you're": "you are",
    "they're": "they are"
}

def expand_contractions(text):
    for contraction, expansion in contractions.items():
        text = text.replace(contraction, expansion)
    return text

# Example
text = "I can't believe it's working and you're happy."
expanded_text = expand_contractions(text)

print("Original Text:")
print(text)

print("\nExpanded Text:")
print(expanded_text)
````

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Data Preparation & Feature Engineering
df = pd.read_csv("TSLA.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

df['Daily_Range'] = df['High'] - df['Low']
df['Close_Open_Diff'] = df['Close'] - df['Open']
df['MA_5'] = df['Close'].rolling(window=5).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()
df['Target_Next_Close'] = df['Close'].shift(-1)
df = df.dropna().reset_index(drop=True)

# Features (X) and Target (y)
feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Daily_Range', 'Close_Open_Diff', 'MA_5', 'MA_20']
X, y = df[feature_cols], df['Target_Next_Close']

# Chronological Split (80% Train, 20% Test)
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# 2. Hyperparameter Optimization & Fitting
# Using TimeSeriesSplit to respect chronological sequence and prevent look-ahead leakage
tscv = TimeSeriesSplit(n_splits=3)
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42), 
    param_grid=param_grid, 
    cv=tscv, 
    scoring='neg_mean_absolute_error', 
    n_jobs=-1
)

print("Optimizing and Fitting the Algorithm...")
grid_search.fit(X_train, y_train)
print(f"Best Hyperparameters: {grid_search.best_params_}")

# 3. Predict on the Model & Evaluate
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calculate and Output Metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
r2 = r2_score(y_test, y_pred)

print(f"\n--- Model Evaluation ---")
print(f"MAE:  ${mae:.2f}")
print(f"RMSE: ${rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
print(f"R²:   {r2:.4f}")
````

##### Which hyperparameter optimization technique have you used and why?

I used GridSearchCV (Grid Search Cross-Validation) combined with TimeSeriesSplit.

* Why GridSearchCV? It exhaustively tests every possible combination of your specified parameters (like `max_depth` and `n_estimators`). For smaller search spaces, it guarantees finding the absolute best tuning configuration without missing optimal values.
* Why TimeSeriesSplit? Stock data is chronological. Standard cross-validation shuffles data, which introduces look-ahead bias (using future prices to predict past prices). `TimeSeriesSplit` creates a rolling window that trains only on the past and validates on the future, accurately simulating real-world trading conditions.

The Impact: This combination successfully optimized the Random Forest model, slashing the forecasting error (MAE) on unseen test data from $38.37 down to $15.83.

````python
# Save the File
import pandas as pd
import numpy as np
import pickle

# =====================================================================
# 1. CUSTOM TREE & RANDOM FOREST ALGORITHM FROM SCRATCH
# =====================================================================

class DecisionTreeRegressorScratch:
    """A single decision tree regressor targeting MSE reduction."""
    def __init__(self, max_depth=5, min_samples_split=5, current_depth=0):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.current_depth = current_depth
        self.left = None
        self.right = None
        self.split_feature = None
        self.split_value = None
        self.value = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.value = np.mean(y)
        
        # Base case stopping criteria
        if (self.current_depth >= self.max_depth or 
            num_samples < self.min_samples_split or 
            np.all(y == y[0])):
            return self

        best_mse = float('inf')
        
        # Find the best split across features and thresholds
        for feature in range(num_features):
            feature_values = X[:, feature]
            thresholds = np.unique(feature_values)
            
            # Subsample unique values if too large to speed up scratch execution
            if len(thresholds) > 20:
                thresholds = np.percentile(feature_values, np.linspace(5, 95, 10))

            for threshold in thresholds:
                left_mask = feature_values <= threshold
                right_mask = ~left_mask
                
                if not np.any(left_mask) or not np.any(right_mask):
                    continue
                
                # Calculate combined MSE of the split
                left_y, right_y = y[left_mask], y[right_mask]
                current_mse = (np.sum((left_y - np.mean(left_y))**2) + 
                               np.sum((right_y - np.mean(right_y))**2))
                
                if current_mse < best_mse:
                    best_mse = current_mse
                    self.split_feature = feature
                    self.split_value = threshold

        # Split nodes recursively if a valid partition is found
        if self.split_feature is not None:
            left_mask = X[:, self.split_feature] <= self.split_value
            right_mask = ~left_mask
            
            self.left = DecisionTreeRegressorScratch(self.max_depth, self.min_samples_split, self.current_depth + 1)
            self.right = DecisionTreeRegressorScratch(self.max_depth, self.min_samples_split, self.current_depth + 1)
            
            self.left.fit(X[left_mask], y[left_mask])
            self.right.fit(X[right_mask], y[right_mask])
            
        return self

    def predict_row(self, row):
        if self.split_feature is None or self.left is None:
            return self.value
        if row[self.split_feature] <= self.split_value:
            return self.left.predict_row(row)
        return self.right.predict_row(row)

    def predict(self, X):
        return np.array([self.predict_row(row) for row in X])


class RandomForestRegressorScratch:
    """An Ensemble Random Forest Model using custom Bootstrap Aggregating."""
    def __init__(self, n_estimators=50, max_depth=5, min_samples_split=5):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.trees = []

    def fit(self, X, y):
        self.trees = []
        num_samples = X.shape[0]
        
        for _ in range(self.n_estimators):
            # Bootstrap Sampling: Random choice with replacement
            indices = np.random.choice(num_samples, size=num_samples, replace=True)
            X_boot, y_boot = X[indices], y[indices]
            
            # Instantiating and training a single tree
            tree = DecisionTreeRegressorScratch(max_depth=self.max_depth, min_samples_split=self.min_samples_split)
            tree.fit(X_boot, y_boot)
            self.trees.append(tree)
        return self

    def predict(self, X):
        # Average the predictions across all internal ensemble trees
        tree_preds = np.array([tree.predict(X) for tree in self.trees])
        return np.mean(tree_preds, axis=0)


# =====================================================================
# 2. DATA PREPARATION & CHRONOLOGICAL TRAIN-TEST SPLIT
# =====================================================================

# Load historical TSLA metrics
df = pd.read_csv("TSLA.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Feature Engineering
df['Daily_Range'] = df['High'] - df['Low']
df['Close_Open_Diff'] = df['Close'] - df['Open']
df['MA_5'] = df['Close'].rolling(window=5).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()

# Target Variable setup
df['Target_Next_Close'] = df['Close'].shift(-1)
df = df.dropna().reset_index(drop=True)

# Define X (Features Matrix) and y (Target Vector)
feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Daily_Range', 'Close_Open_Diff', 'MA_5', 'MA_20']
X_raw = df[feature_cols].values
y_raw = df['Target_Next_Close'].values

# Chronological Train-Test Split (80% Train, 20% Holdout Test)
split_idx = int(len(df) * 0.8)
X_train, X_test = X_raw[:split_idx], X_raw[split_idx:]
y_train, y_test = y_raw[:split_idx], y_raw[split_idx:]


# =====================================================================
# 3. HYPERPARAMETER GRID TUNING & MODEL FITTING
# =====================================================================

print("Running manual parameter exploration over chronological windows...")

# Simulated Hyperparameter Parameter Sets
param_candidates = [
    {'n_estimators': 15, 'max_depth': 4, 'min_samples_split': 5},
    {'n_estimators': 30, 'max_depth': 6, 'min_samples_split': 5}
]

best_mae = float('inf')
best_rf_model = None

# Custom chronological validation logic (Time Series Split representation)
cv_split = int(len(X_train) * 0.75)
X_cv_train, X_cv_val = X_train[:cv_split], X_train[cv_split:]
y_cv_train, y_cv_val = y_train[:cv_split], y_train[cv_split:]

for params in param_candidates:
    # Train candidate architecture
    rf_candidate = RandomForestRegressorScratch(**params)
    rf_candidate.fit(X_cv_train, y_cv_train)
    
    # Check error over validation timeframe
    val_preds = rf_candidate.predict(X_cv_val)
    val_mae = np.mean(np.abs(y_cv_val - val_preds))
    
    if val_mae < best_mae:
        best_mae = val_mae
        # Retrain optimal parameters on the full training array
        best_rf_model = RandomForestRegressorScratch(**params)
        best_rf_model.fit(X_train, y_train)

print(f"Algorithm successfully fitted on complete training set.")


# =====================================================================
# 4. PREDICT ON UNSEEN DATA & EVALUATE
# =====================================================================

print("Generating final forecasts on unseen test horizon...")
y_pred = best_rf_model.predict(X_test)

# Custom performance metric loops
errors = y_test - y_pred
mae = np.mean(np.abs(errors))
rmse = np.sqrt(np.mean(errors ** 2))
mape = np.mean(np.abs(errors / y_test)) * 100

print(f"\n--- From-Scratch Model Evaluation Metrics ---")
print(f"MAE:  ${mae:.2f}")
print(f"RMSE: ${rmse:.2f}")
print(f"MAPE: {mape:.2f}%")


# =====================================================================
# 5. SAVE PERFORMANCE-OPTIMIZED MODEL ATTACHMENT
# =====================================================================

pickle_filename = 'tsla_optimized_rf_model.pkl'
with open(pickle_filename, 'wb') as file:
    pickle.dump(best_rf_model, file)

print(f"\nModel successfully serialized and saved as: {pickle_filename}")
````
